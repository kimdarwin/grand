```python
import json
import shutil
qfile = "q2num.jsonl";
afile = 'a2num.jsonl';
outfile = 'problemsheet2num.json'
out = open(outfile,'w',encoding='utf-8');

qin = open(qfile,'r',encoding='utf-8');
qlines = qin.readlines();
qin.close();

qin = open(afile,'r',encoding='utf-8');
alines = qin.readlines();
qin.close();

outjs=[]
i = -1;
for qline in qlines:
    i = i+1;
    if(i%10==0):
        print(str(i))
    q = json.loads(qline)
    #para = q['text']
    annos = q['annotations']
    m = q['meta']
    para = m['sent']
    sent = q['text']
    problem_no = str(m['problem_no'])
    doc_file = m['doc_file']
    try:
        shutil.copy("../report_txt/"+doc_file,"../report_txt_43/"+doc_file)
    except:
        pass;
    
    if(len(annos)==0):
        continue;
    question = annos[0]['text']
    
    a = json.loads(alines[i]);
    atext = a['text'];
    labels = a['labels'];
    if(len(labels)==0):
        continue;
    answer_form = []
    
    #print(str(labels))
    #sorted(labels, key=lambda x:x[0])
    labels.sort()
    #print(str(labels))
    #print("\n")
    
    #answer_form 생성 시작
    topic = ""
    qpoint = ""
    objects = []
    answers = []
    for label in labels:
        ####form = {}
        start = label[0];
        end = label[1];
        #l = atext[start:end]
        l = atext[start:end].strip()
        t = label[2];
        
        if(t=='topic'):
            topic = l
        elif(t=='qpoint'):
            qpoint = qpoint+" "+l
        elif(t=='object'):
            objects.append(l);
        elif(t=='answer'):
            answers.append(l);
        ####form['answer']=answer
        ####answer_form.append(form)
    
    qpoint = qpoint.strip();
    j = -1;
    for answer in answers:
        j = j+1;
        f = {}
        try:
            f[topic]=objects[j];
        except:
            pass;
        f[qpoint]="";
        answer_form.append(f);
        
    #sorted(answer_form, key = lambda x:x[')
    outj = {}
    #outj['problem_no']=str(i)
    outj['problem_no']=problem_no
    outj['task_no']="2"
    #outj['doc_file']=str(i)
    outj['doc_file']=doc_file
    outj['answer_form']=answer_form
    outj['answers'] = answers
    outj['question']=question
    outj['para'] = para
    
    outjs.append(outj)
out.write(json.dumps(outjs,ensure_ascii=False, indent=4))
out.close();


```
